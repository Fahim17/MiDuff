```python

import os
from PIL import Image
import matplotlib.pyplot as plt

exp_id = "51434428" #"51415276" #"51434428"
idx = 150
# Paths
folder_Y = f"/home/fahimul/Documents/Research/MIDuff/output/{exp_id}"
# folder_DR = "/home/fahimul/Documents/Research/Dataset/University-1651/test/query_drone"
# folder_ST = "/home/fahimul/Documents/Research/Dataset/University-1651/test/query_street"

folder_DR = "/home/fahimul/Documents/Research/Dataset/University-1651/test/query_drone"
folder_ST = "/home/fahimul/Documents/Research/Dataset/University-1651/test/query_satellite"


# Get the 5 image filenames from Y
image_names = [f for f in os.listdir(folder_Y) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

# Sort or slice to ensure only 5 if needed
image_names = image_names[idx:idx+5]

# Plot
fig, axs = plt.subplots(len(image_names), 3, figsize=(8, 3 * len(image_names)))

for i, name in enumerate(image_names):
    # Load image from Y
    img_y = Image.open(os.path.join(folder_Y, name))


    # Corresponding image in X/name/
    name_wo_ext = os.path.splitext(name)[0]
    subfolder_path1 = os.path.join(folder_DR, name_wo_ext)
    subfolder_path2 = os.path.join(folder_ST, name_wo_ext)

    
    # Find image file in subfolder
    x_images = [f for f in os.listdir(subfolder_path1) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
    st_images = [f for f in os.listdir(subfolder_path2) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    
    if not x_images:
        raise FileNotFoundError(f"No image found in {subfolder_path1}")
    
    img_x = Image.open(os.path.join(subfolder_path1, x_images[0]))
    img_st = Image.open(os.path.join(subfolder_path2, st_images[0]))


    # Display images


    axs[i, 0].imshow(img_st)
    axs[i, 0].set_title(f"GT-Sat: {name_wo_ext}.jpeg")
    axs[i, 0].axis('off')

    axs[i, 1].imshow(img_x)
    axs[i, 1].set_title(f"GT-Drone: {name_wo_ext}.jpeg")
    axs[i, 1].axis('off')

    axs[i, 2].imshow(img_y)
    axs[i, 2].set_title(f"Generated: {name}")
    axs[i, 2].axis('off')

plt.tight_layout()
plt.show()
```
